# Example 1: Space Biology to Therapeutics

This notebook demonstrates how to:
1. Get differentially expressed genes from GeneLab
2. Enrich with PrimeKG drug targets
3. Add geospatial context from SPOKE-OKN

```python
from mcp_space_life_sciences import IntegratedKGClient

client = IntegratedKGClient()
```

```python
# Step 1: Get DE genes from GeneLab
de_genes = client.get_genelab_de_genes(
    assay_id="OSD-253-EXAMPLE",
    log2fc_threshold=1.0
)

print(f"Found {len(de_genes['upregulated'])} upregulated genes")
```

```python
# Step 2: Enrich with PrimeKG
enrichment = client.enrich_genes_with_primekg(
    gene_names=de_genes['upregulated'][:50]
)

print(f"Found {len(enrichment['drugs'])} drug targets")
```

```python
# Step 3: Add geospatial context
prevalence = client.get_disease_prevalence_by_location(
    disease_names=enrichment['diseases'][:10],
    location="United States"
)

print(prevalence)
```
